# Что такое разметка?

Разметка аудио это расшифровка слов, которые произносятся на аудио. Помимо слов мы считаем необходимым указывать все междометия, которые так или иначе произносит человек. Ведь любое "ага" или "угу" является полноценной заменой полноценного слова "Да" и передают информацию о согласии. 

# Запись неоднозначных случаев

Как записать иностранное слово? Как записать число? Как записать фамильярное "Здрасьте"? И т.п.

Ответов на эти и многие другие вопросы относительно записи тех или иных сложных случаев однозначных нет. Поэтому в нашей разметке мы придерживаемся правила: лучше — больше. Если записать цифру пять цифрой, то будет потеряна информация о том, какое слово было произнесено "пять", "пятерка", "пятого" и т.д. Поэтому для записи чисел и цифр мы не заменяем слово на цифру. Ниже мы перечислим некоторые простые правила:

* Цифры: пятёрка, пять, пятого — 5
* Иностранные слова: лейтер, лейта, лэйтэ — later
* Буква Ё: ещё — еще
* Спец символы: яндекс точка ру — yandex.ru
* Междометия: ааа, угу, ага — 
* Неразборчивые слова: ?

Более сложные ситуации:
1. Жаргонизмы: здрасьте, здрасте — здравствуйте

    Часто человек не слух не может понять слово было сказано намеренно в разговорной форме или это проблемы со звуком. Поэтому, слово "щас" может быть записано как "щас" или как "сейчас". Чем лучше понятно, что было произнесено в разговорной форме, тем больше вероятность увидеть "щас". Но если было сказано слово "сейчас", то "щас" в разметке записано не будет.
    
2. Аббревиатуры, формулы и неизвестные слова
   Чтобы минимизировать неоднозначность и получить максимум информации, аббревиатуры, формулы и просто неизвестные слова мы прописим записывать исполнителя так, как они слышатся. Такое представление, при желании, можно привести обратно к виду формулы или аббревиатуры.
   
    * Аббревиатуры: "МГУ" правильно записать [эм гэ у].
    * Формулы: "H2O" записывается как [аш два о].

# Защита персональных данных

Аудиозаписи могут быть очень разными по длительности и содержанию. Мы заинтересованы в том, чтобы личные данные людей, которые могут появиться на записи не дошли до случайного исполнителя в первозданном виде. Поэтому прежде чем запись будет отправлена на расшифровку мы предпринимаем обработку со слудующими стадиями:

1. аудио разрезается на небольшие фрагменты со сдвигом мньшим, чем размер фрагмента (так каждый промежуток аудио будет присутствовать сразу в нескольких фрагментах)
2. голос на аудио меняется случайным образом до неузноваемости
3. после разметки фрагменты склеиваются обратно, образуя текст, соотвтетствующий всему исходному аудио

Размер фрагмента выбран таким образом, чтобы минимизировать вероятность попадания персональных данных, а с другой не лишить поностью смысла текста на фрагменте.

# Особенности

Почему алгоритм работает так долго?

Разметка производится набором людей, а потому скорость зависит от количества доступных исполнителей. Также на скорость влияет качество и количество аудио. Сложные записи размечать труднее, следовательно времени исполнителям требуется больше. Если суммарная длительность загруженных на разметку аудио будет слишком большой, то разметка может занять более суток. В этом случае в ночное время может наблюдаться снижение активности исполнителей. Принимая во внимание все эти факторы, мы не можем гарантировать, что 10 часов аудио будет размечено за время в 10 раз большее, чем 1 час. Поэтому рекомендуем загружать на разметку аудио порциями, длительность каждой из которых  с суммарной длительностью не более 10 часов.
